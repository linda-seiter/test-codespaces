```python
'''

Old MacDonald had a farm, E-I-E-I-O!
And on his farm he had a cow, E-I-E-I-O!
With a moo-moo here and a moo-moo there,
Here a moo, there a moo,
Everywhere a moo-moo,
Old MacDonald had a farm, E-I-E-I-O!

Old MacDonald had a farm, E-I-E-I-O!
And on his farm he had a pig, E-I-E-I-O!
With a oink-oink here and a oink-oink there,
Here a oink, there a oink,
Everywhere a oink-oink,
Old MacDonald had a farm, E-I-E-I-O!


'''

#Function parameter animal and sound
def old_macdonald_verse(animal, sound): 
    print('Old MacDonald had a farm, E-I-E-I-O!')
    print('And on his farm he had a {}, E-I-E-I-O!'.format(animal))
    print('With a {}-{} here and a {}-{} there,'.format(sound, sound, sound, sound))
    print('Here a {}, there a {},'.format(sound, sound))
    print('Everywhere a {}-{},'.format(sound, sound))
    print('Old MacDonald had a farm, E-I-E-I-O!')



#main algorithm

#call the function with argument values cow and moo
old_macdonald_verse('cow','moo')  


#call the function with argument values pig and oink
old_macdonald_verse('pig','oink')  

```

```python
import math
import random

#Parameter variables size, diameter, price are initialized when function is called
#Parameter variables are only accessible within the function
def costPerInch( size,  diameter,  price) :
    
    print("Calculating cost for " ,size, " pizza with diameter " , diameter , " and price " , price)
    
    #Local variables radius, area, costPerSqIn, crust_circumf
    #Local variables are only accessible within the function
    radius = diameter / 2
    area = math.pi * math.pow(radius  , 2)
    print("Area = " , area)
    
    costPerSqIn = price / area
    print("Cost per sq inch = " , costPerSqIn)
    
    crust_circumf = 2 * math.pi * radius
    print("Crust circumference = ", crust_circumf)
         

#main algorithm
costPerInch("small", 10, 9.99)  
costPerInch("medium", 12, 12.99)
costPerInch("large", 15, 15.99)

#It is an error to access local variable or parameter variable outside of the function
#print(size)
#print(radius)
```

```python
#Challenge.  Refactor this code to create a new function calculate_cost

weight1 = 22.0
weight2 = 10.0
weight3 = 12.0

#calculate cost for item#1
if weight1 < 15.0 :
    cost1 = 9.95
else :
    cost1 = 12.95
print('cost for {} is {}'.format(weight1,cost1))

#calculate cost for item#2
if weight2 < 15.0 :
    cost2 = 9.95
else :
    cost2 = 12.95
print('cost for {} is {}'.format(weight2,cost2))


#calculate cost for item#3
if weight3 < 15.0 :
    cost3 = 9.95
else :
    cost3 = 12.95
print('cost for {} is {}'.format(weight3,cost3))

```

```python
#Solution

def calculate_cost(weight):
    if weight < 15.0 :
        cost = 9.95
    else :
        cost = 12.95
    print('cost for {} is {}'.format(weight,cost))


#main algorithm
weight1 = 22.0
weight2 = 10.0
weight3 = 12.0

calculate_cost(weight1)
calculate_cost(weight2)
calculate_cost(weight3)


```
